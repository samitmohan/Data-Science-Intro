```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl_df = pd.read_csv("matches.csv")
ipl_df.head()
```

```python
sns.barplot(x = "season", y = "city", data = ipl_df) 
```

```python
plt.hist(ipl_df['win_by_wickets']) 
```

```python
sns.distplot(ipl_df['win_by_wickets'])
```

```python
plt.boxplot(ipl_df['win_by_wickets'])
```

```python
sns.regplot(x = "season", y = "city", data = ipl_df)
```

```python
features = ["season", "win_by_wickets", "city"]
sns.pairplot(ipl_df[features], size = 2)
```

```python
sns.heatmap(ipl_df[features].corr(), annot = True)
```
